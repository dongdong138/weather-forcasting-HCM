```python
f = open("demofile3.txt", "r")
date = []
df_raw = []
for i in f.readlines():
    date.append(i[:10])
    df_raw.append(i[11:])
```

```python
import pandas as pd
df = pd.DataFrame(columns=['Date', 'Time', 'Temperature', 'Dew Point', 'Humidity', 'Pressure', 'Wind Speed', 'wind', 'uv_desc', 'uv_index'])
```

```python
match_dict = {"temp":'Temperature', "dewPt":'Dew Point', "rh":'Humidity', "pressure":'pressure', "wspd":'Wind Speed',
              "wdir_cardinal":'wind', "uv_desc":'uv_desc', "uv_index":'uv_index'}
```

```python
for i in range(len(date)):
    day_df = eval(df_raw[i])
    day_df=day_df['observations']
    for j in day_df:
        a = [[date[i], str(round((j['valid_time_gmt']+ i*86400 -time_root)/3600, 2)), j['temp'], j['dewPt'], j['rh'], j['pressure'], j['wspd'], j['wdir_cardinal'], j['uv_desc'], j['uv_index']]]
        df2 = pd.DataFrame(a,
                            columns=['Date', 'Time', 'Temperature', 'Dew Point', 'Humidity', 'Pressure', 'Wind Speed', 'wind', 'uv_desc', 'uv_index'])
        df=pd.concat([df, df2])
        print(i)
```
